```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
%matplotlib inline
```

```python
# Create a heatmap to visualize the relationships between the numeric variables

# Exclude the non-numeric column (the 'location' cloumn in this case)
numeric_df = df.select_dtypes(include=['int', 'float'])

# Create the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
```

### Interpretations of the result

1. TotalDeathPercentage vs. TotalVaccinationPercentage:
- Correlation coefficient: -0.24
- Interpretation: There is a weak negative correlation between the percentage of total deaths and the percentage of total vaccinations. This suggests that countries with higher vaccination percentages tend to have slightly lower death percentages, and vice versa.

2. TotalDeathPercentage vs. ExtremePovertyPercentage:
- Correlation coefficient: 0.09
- Interpretation: There is a weak positive correlation between the percentage of total deaths and the percentage of extreme poverty. This suggests that countries with higher percentages of extreme poverty tend to have slightly higher death percentages, and vice versa.

3. TotalDeathPercentage vs. GDPPerCapita:

4. TotalDeathPercentage vs. Aged70OverPercentage:

- Interpretation: There is a weak negative correlation between the percentage of total deaths and the percentage of population aged 70 and over. This suggests that countries with higher proportions of elderly populations tend to have slightly lower death percentages, and vice versa.

7. TotalDeathPercentage vs. DiabetesPrevalence:

The above interpretations provide insights into the relationships between the variables based on the correlation coefficients. The correlation does not imply causation, and other factors may influence the observed relationships.
